```python
# LINEAR REGRESSION USING SCIKIT LEARN

from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris = load_iris()

print(iris)
```

```python
x = pd.DataFrame(data= iris.data, columns= iris.feature_names)
x
```

```python
y = pd.DataFrame(data= iris.target, columns= ['species'])
y
```

```python
data=pd.concat([x, y], axis= 1)
print(data)
```

```python
y1= data['sepal length (cm)']
y1
```

```python
x1= data.iloc[:,1:5]
x1
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=0.3,random_state=42)

print(x_train)
print(x_train.loc[81])
#print(x_test.loc[81])
print(data.loc[81])
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
print(lr)
```

```python
lr.fit(x_train,y_train)
print(x_test)
```

```python
y_predict = lr.predict(x_test)
y_predict
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
print('Mean Absolute Error:', mean_absolute_error(y_test, y_predict))
print('Mean Squared Error:', mean_squared_error(y_test,y_predict))
```

```python
y_predict= lr.predict ([[2.8,4.7,1.2,1]])
print(y_predict)
```

```python
y_predict= lr.predict ([[2.4,3.7,1.0,1.0]])
print(y_predict)
```

```python
# LOGISTIC REGRESSION
iris= load_iris()
iris
```

```python
x = pd.DataFrame(data= iris.data, columns= iris.feature_names)
x
```

```python
y = pd.DataFrame(data= iris.target, columns= ['species'])
y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)

print(x_train)
```

```python
from sklearn.model_selection import KFold   
k_value = 6  
k_fold = KFold(n_splits = k_value, random_state = None)    
```

```python
from sklearn.linear_model import LogisticRegression
lo=LogisticRegression()
print(lo)
```

```python
lo.fit(x_train,y_train)
print(x_test)

from sklearn.model_selection import cross_val_score  
scores = cross_val_score(lo, x_train, y_train, cv = k_fold)  
print("Accuracy score of each fold: ", scores)
```

```python
y_predict = lo.predict(x_test)
y_predict
```

```python
from sklearn.metrics import accuracy_score
accuracy= accuracy_score(y_test,y_predict)
print(accuracy)
```
